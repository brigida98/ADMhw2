Write another function that, given an input n (an integer), will return the posts that belong to the n top posted profiles (top n profiles that have posted the highest number of posts) that their data is available in the profile.csv using the previously written function.

```python
n_top = int(input('Enter a Number:'))
```

```python
#sort based on the number of posts a user shared (profile.csv)
data_profile.sort_values(by = 'n_posts' , ascending = False, inplace = True)
```

```python
#create a sub-dataframe of n-top users (profile.csv)
selected = data_profile.head(n_top)
```

```python
#merge 2 dataframes on profile_id column (profile.csv and post.csv)
merged = pd.merge(selected['profile_id'], data_post[['profile_id','post_id']], on = 'profile_id', how = "left")
```

What is the average number of "likes" and comments of the top 10 profiles with the highest number of posts which their information is available in profile.csv?

```python
#create a sub-dataframe contains top 10 profiles data (profile.csv)
selected_2 = data_profile.sort_values(by = 'n_posts', ascending = False).head(10)
```

```python
merged_2 = pd.merge(selected_2[['profile_id','followers','following']], data_post, on = "profile_id", how = "left")
```

```python
print('average number of likes: ' ,merged_2['numbr_likes'].mean(axis = 0, skipna = True).round(2))
```

```python
print('average number of comments: ' ,merged_2['number_comments'].mean(axis = 0, skipna = True).round(2))
```
